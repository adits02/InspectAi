```python
class IndianClassroomNormsAnalyzer:
    def __init__(self):
        # Detailed geographical and climate zones of India
        self.regions = {
            'Northern Plains': {
                'states': ['Delhi', 'Uttar Pradesh', 'Haryana', 'Punjab', 'Rajasthan'],
                'climate': 'Extreme (Hot Summers, Cold Winters)',
                'key_challenges': [
                    'Temperature Variation',
                    'High Solar Radiation',
                    'Dust Protection'
                ],
                'classroom_norms': {
                    'window_orientation': 'North-South',
                    'recommended_window_area': (0.15, 0.20),  # 15-20% of floor area
                    'ceiling_height': (3.0, 3.5),  # meters
                    'ventilation_rate': (4, 6),  # air changes per hour
                    'thermal_comfort_temp': (24, 26),  # degrees Celsius
                }
            },
            'Western Coastal': {
                'states': ['Maharashtra', 'Goa', 'Gujarat'],
                'climate': 'Humid Tropical',
                'key_challenges': [
                    'High Humidity',
                    'Monsoon Rains',
                    'Corrosion Risk'
                ],
                'classroom_norms': {
                    'window_orientation': 'East-West with overhangs',
                    'recommended_window_area': (0.20, 0.25),  # 20-25% of floor area
                    'ceiling_height': (3.5, 4.0),  # meters
                    'ventilation_rate': (6, 8),  # air changes per hour
                    'thermal_comfort_temp': (23, 25),  # degrees Celsius
                }
            },
            'Southern Plateau': {
                'states': ['Karnataka', 'Andhra Pradesh', 'Tamil Nadu', 'Telangana'],
                'climate': 'Semi-Arid to Tropical',
                'key_challenges': [
                    'Temperature Fluctuations',
                    'Glare Management',
                    'Dust Control'
                ],
                'classroom_norms': {
                    'window_orientation': 'Northeast-Southwest',
                    'recommended_window_area': (0.18, 0.22),  # 18-22% of floor area
                    'ceiling_height': (3.2, 3.7),  # meters
                    'ventilation_rate': (5, 7),  # air changes per hour
                    'thermal_comfort_temp': (24, 26),  # degrees Celsius
                }
            },
            'Eastern Coastal': {
                'states': ['Odisha', 'West Bengal', 'Andhra Pradesh'],
                'climate': 'Tropical Monsoon',
                'key_challenges': [
                    'High Rainfall',
                    'Humidity',
                    'Cyclone Prone Areas'
                ],
                'classroom_norms': {
                    'window_orientation': 'East with deep overhangs',
                    'recommended_window_area': (0.22, 0.27),  # 22-27% of floor area
                    'ceiling_height': (3.6, 4.1),  # meters
                    'ventilation_rate': (7, 9),  # air changes per hour
                    'thermal_comfort_temp': (23, 25),  # degrees Celsius
                }
            },
            'Himalayan Region': {
                'states': ['Himachal Pradesh', 'Uttarakhand', 'Jammu & Kashmir'],
                'climate': 'Alpine to Temperate',
                'key_challenges': [
                    'Cold Temperature',
                    'Limited Sunlight',
                    'Snow Protection'
                ],
                'classroom_norms': {
                    'window_orientation': 'South-facing',
                    'recommended_window_area': (0.15, 0.20),  # 15-20% of floor area
                    'ceiling_height': (3.0, 3.5),  # meters
                    'ventilation_rate': (3, 5),  # air changes per hour
                    'thermal_comfort_temp': (22, 24),  # degrees Celsius
                }
            },
            'North-Eastern Hilly': {
                'states': ['Assam', 'Manipur', 'Meghalaya', 'Nagaland', 'Tripura'],
                'climate': 'Sub-Tropical Monsoon',
                'key_challenges': [
                    'High Rainfall',
                    'Humidity',
                    'Terrain Variations'
                ],
                'classroom_norms': {
                    'window_orientation': 'Northeast with side ventilation',
                    'recommended_window_area': (0.20, 0.25),  # 20-25% of floor area
                    'ceiling_height': (3.3, 3.8),  # meters
                    'ventilation_rate': (5, 7),  # air changes per hour
                    'thermal_comfort_temp': (23, 25),  # degrees Celsius
                }
            }
        }
    
    def generate_comprehensive_report(self):
        """
        Generate a comprehensive report of regional classroom norms
        """
        report_data = []
        
        for region, details in self.regions.items():
            region_report = {
                'Region': region,
                'Climate': details['climate'],
                'Key_Challenges': ', '.join(details['key_challenges']),
                'Window_Orientation': details['classroom_norms']['window_orientation'],
                'Min_Window_Area_Percentage': details['classroom_norms']['recommended_window_area'][0] * 100,
                'Max_Window_Area_Percentage': details['classroom_norms']['recommended_window_area'][1] * 100,
                'Min_Ceiling_Height_m': details['classroom_norms']['ceiling_height'][0],
                'Max_Ceiling_Height_m': details['classroom_norms']['ceiling_height'][1],
                'Min_Ventilation_Rate': details['classroom_norms']['ventilation_rate'][0],
                'Max_Ventilation_Rate': details['classroom_norms']['ventilation_rate'][1],
                'Min_Thermal_Comfort_Temp': details['classroom_norms']['thermal_comfort_temp'][0],
                'Max_Thermal_Comfort_Temp': details['classroom_norms']['thermal_comfort_temp'][1]
            }
            report_data.append(region_report)
        
        # Convert to DataFrame
        df = pd.DataFrame(report_data)
        
        # Save to Excel
        df.to_excel('indian_classroom_regional_norms.xlsx', index=False)
        
        return df
    
    def calculate_comfort_index(self, region, actual_measurements):
        """
        Calculate comfort index based on regional norms
        """
        if region not in self.regions:
            raise ValueError(f"Region {region} not found in database")
        
        norms = self.regions[region]['classroom_norms']
        comfort_scores = {}
        
        # Window Area Comfort
        window_area_score = self._calculate_score(
            actual_measurements['window_area_percentage'],
            norms['recommended_window_area'][0] * 100,
            norms['recommended_window_area'][1] * 100
        )
        comfort_scores['window_area_comfort'] = window_area_score
        
        # Ceiling Height Comfort
        ceiling_height_score = self._calculate_score(
            actual_measurements['ceiling_height'],
            norms['ceiling_height'][0],
            norms['ceiling_height'][1]
        )
        comfort_scores['ceiling_height_comfort'] = ceiling_height_score
        
        # Ventilation Comfort
        ventilation_score = self._calculate_score(
            actual_measurements['ventilation_rate'],
            norms['ventilation_rate'][0],
            norms['ventilation_rate'][1]
        )
        comfort_scores['ventilation_comfort'] = ventilation_score
        
        # Thermal Comfort
        thermal_score = self._calculate_score(
            actual_measurements['thermal_comfort_temp'],
            norms['thermal_comfort_temp'][0],
            norms['thermal_comfort_temp'][1]
        )
        comfort_scores['thermal_comfort'] = thermal_score
        
        # Overall Comfort Index
        comfort_scores['overall_comfort_index'] = np.mean(list(comfort_scores.values()))
        
        return comfort_scores
    
    def _calculate_score(self, actual_value, min_value, max_value):
        """
        Calculate a comfort score based on how close the actual value is to the recommended range
        """
        if actual_value < min_value:
            return max(0, 10 - (min_value - actual_value))
        elif actual_value > max_value:
            return max(0, 10 - (actual_value - max_value))
        else:
            # Perfect score if within range
            return 10
```

```python
def main():
    # Initialize the analyzer
    classroom_norms = IndianClassroomNormsAnalyzer()
    
    # Generate comprehensive report
    regional_report = classroom_norms.generate_comprehensive_report()
    print("Regional Classroom Norms Report Generated!")
    
    # Example comfort index calculation
    try:
        comfort_index = classroom_norms.calculate_comfort_index(
            'Southern Plateau', 
            {
                'window_area_percentage': 20,  # percentage of floor area
                'ceiling_height': 3.5,  # meters
                'ventilation_rate': 6,  # air changes per hour
                'thermal_comfort_temp': 25  # degrees Celsius
            }
        )
        
        print("\nComfort Index Analysis:")
        for key, value in comfort_index.items():
            print(f"{key.replace('_', ' ').title()}: {value:.2f}/10")
    
    except Exception as e:
        print(f"Error in comfort index calculation: {e}")

if __name__ == "__main__":
    main()
```
