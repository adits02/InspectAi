```python
import json
from textblob import TextBlob
from transformers import pipeline

def perform_sentiment_analysis(text):
    """Perform sentiment analysis on the given text."""
    blob = TextBlob(text)
    sentiment = blob.sentiment
    sentiment_result = {
        "polarity": sentiment.polarity,
        "subjectivity": sentiment.subjectivity,
        "sentiment_label": "positive" if sentiment.polarity > 0 else "negative" if sentiment.polarity < 0 else "neutral"
    }
    return sentiment_result

def summarize_text(text):
    """Summarize the given text using a BERT-based extractive summarization model."""
    summarizer = pipeline("summarization")
    try:
        summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
        summary = summary[0]['summary_text']
    except Exception as e:
        print(f"Error in summarization: {e}")
        summary = text  # In case summarization fails
    return summary

def save_to_json(input_text, sentiment, summary, output_file):
    """Save the sentiment analysis and summary into a JSON file in QnA format."""
    data = {
        "Q": input_text,
        "A": {
            "SentimentAnalysis": sentiment,
            "Summary": summary
        }
    }
    with open(output_file, 'w') as json_file:
        json.dump(data, json_file, indent=4)

def main():
    # Example input text
    input_text = """Artificial Intelligence (AI) is transforming the world. It enables machines to learn from data and perform tasks that traditionally required human intelligence. While it brings immense benefits, such as automation and insights, it also raises ethical concerns and challenges."""

    # Perform sentiment analysis
    sentiment = perform_sentiment_analysis(input_text)

    # Summarize the text
    summary = summarize_text(input_text)

    # Save to JSON in QnA format
    output_file = "sentiment_summary_qna.json"
    save_to_json(input_text, sentiment, summary, output_file)

    print(f"Results saved to {output_file}.")

if __name__ == "__main__":
    main()
```
